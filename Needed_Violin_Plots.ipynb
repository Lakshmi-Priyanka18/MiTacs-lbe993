```python
import pandas as pd
import seaborn
data = pd.read_excel(r"E:\Dream\Dream Research\Mitacs Project\RMSD values.xlsx")
print(data.head())

seaborn.set(style = 'whitegrid')

seaborn.violinplot(x="Comparison", y="1H RMSD Distribution / ppm", hue="Data set", data=data,gridsize=100,linewidth=2, width=0.3, inner='box', palette="Set2",cut=0, split=True,scale="count")
```
