```python
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.metrics import mean_squared_error

x=[278,234,232,239,234,246,265,279,279,296,296]
y=[249.124,214.811,210.436,214.609,209.379,227.743,248.654,260.342,264.659,297.652,296.262]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

rmse = mean_squared_error(y, mymodel, squared=False)
plt.title("J-coupling PBE Scalar")
plt.xlabel("Experimental")
plt.ylabel("Computational")
plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
print("R^2 : ",r)
print("Slope : ",slope)
print("Intercept : ",intercept)
print("Std_err : ",std_err)
print("P value : ",p)
print("RMSE : ",rmse)
```
